covariance = sum ( xi-xbar)(yi - ybar) / N

```python
x=np.arange(1,11)
y=np.arange(100,110)
xbar = np.sum(x)/len(x)
ybar=np.sum(y)/len(y)
np.sum((x-xbar)*(y-ybar))/len(y)
```

```python
np.cov(x,y)
```

correlation

```python
std_x=np.sqrt(np.sum(np.square(x-xbar)) / np.count_nonzero(x))
std_y=np.sqrt(np.sum(np.square(y-ybar)) / np.count_nonzero(y))
cor = (np.sum((x-xbar)*(y-ybar))/np.count_nonzero(y)) / (std_x *std_y)
cor
```

cosine similarity = dot(a,b)/norm(a) * norm(b)

```python
x = [1,2,3]
y=[4,5,6]
arr1 = np.array(x)
arr2= np.array(y)
cosine= np.dot(arr1,arr2)/(norm(arr1) * norm(arr2))
cosine
```
