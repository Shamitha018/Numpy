```python
#as python doesnt understand which axis the label 'marks' is present,
#we need to provide axis
#by default it takes axis =0, which means rows so fetch in rows for marks was a fail
#so when u give drop(1) it takes axis 0 that is in rows and drops 1st row
```

```python
df.drop(1)
```

```python
df.drop('Marks', axis = 1) #now it understands to fetch 'marks' in y axis = columns
```

```python
df.drop('Marks', axis=0) #couldnt fetch marks in rows, same as only df.drop('Marks)
```

step 14: creating new df with x=1,10 . y= sqaure of 1,10 , z = cube of 1,10

```python
l1 = [i for i in range(1,11)]
l2 = [i**2 for i in range(1,11)]
l3 = [i**3 for i in range(1,11)]
df2= pd.DataFrame(zip(l1,l2,l3), columns=['x', 'y', 'z'])
df2
```

alternate way with numpy

```python
import numpy as np
df1 = pd.DataFrame()
df1['x']=range(1,11)
df1['y'] = np.square(range(1,11))
df1['z'] = df1['x'] * df1['y']
df1
```
